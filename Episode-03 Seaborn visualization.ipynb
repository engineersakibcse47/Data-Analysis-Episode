```python
import seaborn as sns
import pandas as pd
import numpy as np
from numpy import median
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
fmri = sns.load_dataset('fmri')
df = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
fmri.head()
```

```python
df.head()
```

```python
iris.head()
```

### Line Plot

```python
sns.lineplot(x='timepoint', y='signal', data=fmri)
```

```python
sns.lineplot(x='timepoint', y='signal',
             hue='event',
             style='event',
             markers=True,
             data=fmri)
plt.show()
```

### Bar Plot

```python
sns.barplot(x='day', y='tip', 
            data=df,
            hue= 'sex',
            palette= 'spring', 
            order=['Sat','Sun','Thur','Fri'])
plt.show()
```

*we have heighest tips for sunday*

```python
sns.barplot(x='day', y='total_bill',
            data=df, 
            hue= 'sex', 
            palette= 'husl', # winter_r, hls, Paired,cubehelix, seagreen, autumn
            order=['Sat','Sun','Thur','Fri'],
            estimator=median)
```

### Dist Plot

```python
num= np.random.randn(150)
```

```python
sns.distplot(num,
             color='red',
             rug=True,
             hist=True,
             kde=True)
```

### Box Plot

```python
sns.boxplot(x='day', 
            y='total_bill',
            data=df,
            hue= 'sex',
            palette='autumn',
            order=['Sat','Sun','Thur','Fri']
           )
```

```python
sns.boxplot(x='day', 
            y='total_bill',
            data=df,
            palette='husl'
           )
sns.swarmplot(x='day', 
            y='total_bill',
            data=df,
            palette='autumn'
           )
```

```python
sns.boxplot(data=iris, 
            orient='h',
            palette='cubehelix')
```

```python
sns.boxplot(data=iris, 
            orient='v',
            palette='husl')
```

### Strip Plots

```python
sns.stripplot(x=df['tip'],
             color='seagreen'
             )
```

```python
sns.stripplot(x='day',
              y='total_bill',
              data=df)
```

```python
sns.stripplot(x='day',
              y='total_bill',
              data=df,
              hue='sex'
              )
```

```python
sns.stripplot(x='day',
              y='total_bill',
              data=df,
              hue='sex',
              jitter=False,
              split=True,
              palette='winter_r')
```

```python
sns.stripplot(x='day',
              y='total_bill',
              data=df,
              hue='smoker',
              jitter=False,
              split=True,
              palette='winter_r',
              marker='D')
```

```python
sns.stripplot(x='day',
              y='total_bill',
              data=df,
              palette='autumn',
              jitter=False
              )
sns.boxplot(x='day',
            y='total_bill',
            data=df
              )
```

### Violin Plot

```python
sns.violinplot(x='day',
              y='total_bill',
              data=df,
              hue='sex',
              palette='autumn',
              )
```

```python
sns.violinplot(x='day',
              y='total_bill',
              data=df,
              hue='smoker',
              palette='husl',
              split=True
              )
```

```python
sns.violinplot(x='day',
              y='total_bill',
              data=df,
              hue='smoker',
              palette='winter_r',
              inner='quartile'
              )
```

### FacetGrid

```python
x= sns.FacetGrid(df, row='smoker', col='time')
x= x.map(plt.hist, 'total_bill', color='green', bins=10)
```

```python
x= sns.FacetGrid(df, row='smoker', col='time')
x= x.map(plt.scatter, 'total_bill','tip', color='green')
```

```python
x= sns.FacetGrid(df, row='smoker', col='time', size=4, palette='winter_r')
x= x.map(sns.regplot, 'total_bill','tip', color='seagreen')
```

```python
x= sns.FacetGrid(df, col='time', hue='smoker', size=4, palette='winter_r')
x= (x.map(plt.scatter, 'total_bill','tip').add_legend())
```

```python
x= sns.FacetGrid(df, col='day', size= 5)
x= x.map(sns.boxplot,'time', 'total_bill', color='seagreen')
```

### Joint Plot

```python
sns.jointplot(x='total_bill', 
              y='tip', 
              data=df, 
              kind='reg', 
              color='seagreen')
```

```python
sns.jointplot(x='total_bill', 
              y='tip', 
              data=df, 
              kind='hex', 
              color='seagreen',
              size=7,
              ratio=7)
```

### Regression Plot

```python
sns.regplot(x='total_bill', 
            y='tip', 
            data=df,
            marker= '*', 
            scatter_kws={'color':'green'},
            line_kws={'linewidth':2, 'color':'coral'}
            )
```

```python
sns.regplot(x='size', 
            y='tip', 
            data=df,
            marker= 'D', 
            x_jitter=0.09,
            scatter_kws={'color':'skyblue'},
            line_kws={'linewidth':2, 'color':'coral'}
            )
```

```python
sns.regplot(x='size', 
            y='tip', 
            data=df,
            marker= '*', 
            x_estimator=np.mean,
            scatter_kws={'color':'green'},
            line_kws={'linewidth':2, 'color':'coral'}
            )
```

### Pair Plot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, kind='reg')
```

```python
sns.pairplot(iris, diag_kind='hist', palette='winter_r', hue='species')
```

```python
sns.pairplot(iris, diag_kind='kde',kind='reg', palette='autumn', hue='species')
```

```python
sns.pairplot(iris, hue='species', palette='husl')
```

```python
sns.pairplot(iris, size=4, vars=['sepal_length', 'sepal_width'])
```

```python
sns.pairplot(iris, size=3, 
             x_vars=['sepal_length', 'sepal_width'], 
             y_vars=['petal_length', 'petal_width'],
             palette='husl',
             hue='species')
```
